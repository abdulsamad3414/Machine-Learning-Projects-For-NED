The above box plot shows that there are some outliers in tenure column . We can use the following code to remove outliers from the tenure column:

```python
# Number of rows containing outliers
# Compute the 25th percentile and 75th percentile
q1 = df['tenure'].quantile(0.25)
q3 = df['tenure'].quantile(0.75)

# Compute the interquartile range
iqr = q3 - q1

# Compute the lower and upper bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print("Lower bound: ", lower_bound)
print("Upper bound: ", upper_bound)
print("Number of rows containing outliers: ", df[(df['tenure'] < lower_bound) |(df['tenure'] > upper_bound)].shape[0])
```

# Build the Logistic regression model 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
# encode categorical columns 
le = LabelEncoder()
df["Department"] =le.fit_transform(df["Department"])
df["salary"]=le.fit_transform(df["salary"])
```

```python
df.head(2)
```

```python
df.describe()
```

```python
# need standrad scaling of the average_montly_hours
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df["average_montly_hours"] = scaler.fit_transform(df[["average_montly_hours"]])
df.head(2)
```

```python
df.describe()
```

```python
# Select feature columns 
X = df[["satisfaction_level", "last_evaluation", "number_project", "average_montly_hours", "tenure", "Work_accident", "promotion_last_5years", "Department", "salary"]]
# Select target column
y = df["left"]
```

```python
X.shape
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# Train Model 
LR=LogisticRegression()
LR.fit(X_train,y_train)
```

```python
# model accuracy
from sklearn.metrics import accuracy_score,classification_report
# Predict
y_pred=LR.predict(X_test)
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

# Classification Report
print(classification_report(y_test, y_pred))
```
